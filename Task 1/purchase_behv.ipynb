```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pb = pd.read_csv("QVI_purchase_behaviour.csv")
pb
```

```python
print("Duplicate value:", pb.duplicated().sum())
pb.info()
```

```python
categorical_cols = pb.select_dtypes(include='object')
numeric_cols = pb.select_dtypes(include=['int64', 'float64'])

rows = pb.shape[0]
cols = pb.shape[1]
print(f'the dataset contains {rows} rows and {cols} columns.')

#print the summary statistics for categorical column
print("\nSummary statistics for categorical column:")
display(categorical_cols.describe())

#print the summary statistics for numeric column
print("\nSummary statistics for numeric column:")
display(numeric_cols.describe())
```

```python
features = ['LIFESTAGE', 'PREMIUM_CUSTOMER']
for feature in features:
    print(f"\nColumn: {pb[feature].value_counts()}")
```

```python


for feature in features:
  plt.figure(figsize=(12,4))
  sns.countplot(x = pb[feature], palette = 'viridis', order = pb[feature].value_counts().index)
  plt.title(f'Distribution of {feature}')
  plt.xlabel(feature)
  plt.ylabel('frequency')
  plt.xticks(rotation=45)
  plt.show()
```
